```python
l=[]
nr=1000000
for i in range(nr):
  u=random()
  l.append(sqrt(u))
x1=-1
x2=2
x=x1
b=[]
while x < x2:
  b.append(x)
  x=x + (x2 - x1)/100
plt.hist(l, density=1, bins=b, color="grey", edgecolor="black")
plt.show()
```
