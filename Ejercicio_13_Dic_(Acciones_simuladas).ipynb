```python
from random import gauss
from math import exp, sqrt, log
import matplotlib.pyplot as plt

def precios(p0,u,o,t):
  lp=[]
  p=p0
  for i in range(t):
    p=p*exp(u+o*gauss(0,1))
    lp. append(p)
  return lp
p0=10
u=0.0015
o=0.02
t=30     # dias
lp=precios(p0,u,o,t)
fpt=open("datos.txt", "w")
for i in range(len(lp)):
  fpt. write("%8.3f"%lp[i]+"\n")    # crea lista de valores simulados
fpt. close()
```

```python
import numpy as np
from statistics import mean, variance
from math import log
import matplotlib.pyplot as plt

p=np.loadtxt("datos.txt")      # abrir el archivo
n=len(p)
print(n)
r=np.zeros(n-1)         # rendimientos
for i in range(n-1):
  r[i]=log(p[i+1]/p[i])    # rendimiento ogritmizado
print(len(r))
m=np.mean(r)
o=np.std(r)
print(m,o)
plt. figure(figsize=(3.5,3))
for i in range(5):
  lp=precios(p0,u,o,t)
  plt. plot(lp)
plt. show()
```
