```python
from numba import jit
@jit(nopython=True)
def f(T):
  lam_nac=10
  lam_fal= 9.8
  n=1000
  tn=0
  t=0
  tf=expovariate(lam_fal)
  while min(tn, tf)<T:
    if n<3000:
      lam_nac=n*(3000-n)/100000
    else:
      lam_nac=0
    t=min(tn, tf)
    if t==tn:
      n=n+1
      tn=tn + expovariate(lam_nac)
    else:
      n=n-1
      tf=tf + expovariate(lam_fal)
  return n

nr=100000
l=[]
for i in range(nr):
  l.append((f(200)))

plt.figure(figsize=(3.5, 3))
plt.hist(l, density=1, color="grey", edgecolor="black")
plt.show()
```

```python
import numpy as np
from numba import jit
from statistics import mean#, stdev, variance, media
@jit(nopython=True)
def f(T):
  lam_nac=10
  lam_fal= 9.8
  n=1000
  tn=0
  t=0
  tf=expovariate(lam_fal)
  while min(tn, tf)<T:
    if n<3000:
      lam_nac=n*(3000-n)/100000
    else:
      lam_nac=0
    t=min(tn, tf)
    if t==tn:
      n=n+1
      tn=tn + expovariate(lam_nac)
    else:
      n=n-1
      tf=tf + expovariate(lam_fal)
  return n

nr=100000
l=[]
for i in range(nr):
  l.append((f(200)))
print("Intervalo de confianza del 90% --> [", np.quantile(l, 0.05), ",", np.quantile(l, 0.95), "]")
print("Media = ", mean(l))
#plt.figure(figsize=(3.5, 3))
plt.hist(l, density=1, color="grey", edgecolor="black")
plt.show()
```
