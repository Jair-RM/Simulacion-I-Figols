```python
from random import expovariate
import matplotlib.pyplot as plt

lam_nac=10          # nacimientos ocurridos
lam_fal= 9.8            # fallecimientos ocurridos
n=1000
tn=0
t=0
tf=expovariate(lam_fal)
T=500
lt=[]
lp=[]
while t<T and n<2700 and n>0:
  if n<3000:
    lam_nac=n*(3000-n)/100000
  else:
    lam_nac=0
  lt.append(t)
  lp.append(n)
  t=min(tn, tf)
  lt.append(t)
  lp.append(n)
  if t==tn:
    n=n+1
    tn=tn + expovariate(lam_nac)
    #print("%10.3f"%t, "%6d"%n)
  else:
    n=n-1
    tf=tf + expovariate(lam_fal)
    #print("                                       ", "%10.3f"%t, "%6d"%n)
print(t)
plt.plot(lt, lp)
plt.show()
```
