```python
def fx():
  u=random()
  if u<=0.5:
    x=sqrt(2*u) - 1
  if u>0.5:
    x=1 - sqrt(2-2*u)
  return x

l=[]
N=1000000
for i in range(N):
  l.append(fx())
#print(l)
plt.hist(l, density=1, bins=100000, color="black", edgecolor="purple")
plt.show()
```

```python
# ChatGPT
import numpy as np
import matplotlib.pyplot as plt

# Número de simulaciones
n = 1_000_000

# Generar valores aleatorios uniformes
U = np.random.uniform(0, 1, n)

# Inicializar el array para X
X = np.zeros(n)

# Transformación de U a X según la CDF
# Para la parte de [-1, 0]
mask1 = U < 0.5
X[mask1] = np.sqrt(2 * U[mask1]) - 1

# Para la parte de [0, 1]
mask2 = U >= 0.5
X[mask2] = 1 - np.sqrt(2 * (U[mask2] - 0.5))

# Crear el histograma
plt.hist(X, bins=50, density=True, alpha=0.6, color='g')

# Graficar la función de densidad
x_values = np.linspace(-1, 1, 1000)
pdf_values = np.piecewise(x_values,
                          [x_values < 0, x_values >= 0],
                          [lambda x: x + 1, lambda x: -x + 1])

plt.plot(x_values, pdf_values, 'r', lw=2, label='PDF teórica')
plt.title('Histograma de simulación de la variable aleatoria X')
plt.xlabel('X')
plt.ylabel('Densidad')
plt.legend()
plt.grid()
plt.show()
```

```python
# Codigo profesor

def inversa(x):
  if x>0 and x<1/2:
    return -1 + sqrt(2*x)
  else:
    return 1 - sqrt(2*(1-x))

nr=1000000
l=[]
l1=[]
for i in range (nr):
  x=random()
  l.append(inversa(x))
  z=inversa(x)
  l1.append(sqrt(1+z**2))
print(mean(l1), variance(l1))
x1=-1
x2=1
x=x1
b=[]
while x<x2:
  b.append(x)
  x=x+(x2-x1)/100
plt.figure(figsize=(4,3))
plt.hist(l, density=1, bins=b, color="grey", edgecolor="black")
plt.show()
```
