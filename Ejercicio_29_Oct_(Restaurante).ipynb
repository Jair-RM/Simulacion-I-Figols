En un restaurante hay 3 mesas. Los comensales llegan siguiendo un proceso de Poisson con parametro 6 usuarios/hora. El tiempo que tardan en comer sigue una distibucion exponencial con media 0.5 horas. Si cuando llegan los comensales, las tres mesas estan ocupadas, entonces se forman afuera del restaurante. Estime por simulacion el valor esperado del valor medio del numero de personas afuera, en el intervalo [0,8]

```python
from random import random
from math import log
def imprime(lt):
  n=len(lt)
  print("Tiempos de salida [", end="")
  for i in range(n):
    print("%6.4f"%lt[i], ",", end="")
  print("]")
infi=1000000
l_l=6
t_m=0.5
l_s=1/t_m
mes=3
T=10
l=0
le=0
t=0
t_l=0
lt=[]
t_s=infi
while min(t_s, t_l) < T:
  imprime(lt)
  if len(lt) > 0:
    t_s=min(lt)
  else:
    t_s=infi
  t=min(t_s, t_l)
  if t==t_l:
    if l < mes:
      lt. append(t - (1/l_s)*log(random()))
    t_l=t_l  - (1/l_l)*log(random())
    l=l+1
    if l <= mes:
      le=0
    else:
      le=1 - mes
    print("Llegada ", "%10.4f"%t, "%4d"%l, "%4d"%le)
  else:
    k=lt. index(t)
    lt. pop(k)
    if l > mes:
      lt. append(t - (1/l_s)*log(random()))
    l=l - 1
    if l <= mes:
      le=0
    else:
      le= l - mes
    print("          Salida  ", "%10.4f"%t, "%4d"%l, "%4d"%le)
# imprimir primero el numero de ersonas en el sistema y luego el numero
# < > ∈ o ∉ | λ
```
