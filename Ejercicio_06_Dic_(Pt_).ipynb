```python
# codigo de bondad de ajuste para ver si sigue una distribucion normal
from random import gauss
from math import exp, sqrt
import matplotlib.pyplot as plt
from scipy.stats import norm, kstest

def precio30(p0,u,o,T):
  return p0*exp(u*T + o*sqrt(T)*gauss(0,1))
u=0.001
o=0.02
p0=10
t=30
l=[]
nr=100000
for i in range(nr):
  l. append(precio30(p0,u,o,T))
plt. hist(l, density=1, bins=50)
plt. show()
par=norm.fit(l)
print(par)
ks=kstest(l, "norm", args=(par[0], par[1]))        # kornovorol smirlov
print(ks)
```
