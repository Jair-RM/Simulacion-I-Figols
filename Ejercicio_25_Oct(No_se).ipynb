```python
def demanda():
  u=random.uniform(0,1)
  if u < 0.1:
    d=0
  elif u < 0.5:
    d=1
  elif u < 0.8:
    d=2
  elif u < 0.9:
    d=3
  else:
    d=4
  return d

T=100
s=50
S=150
q=S
l_l=0.5
t=0
tl=random.exponential(l_l)
tr=20
rep=30
v=0
nv=0
costo=0
ganancia=0
costounitario=0.05
gananciaunitaria=30

while min(tl, tr) < T:
  ta=t
  t=min(tl, tr)
  costo=costo + (t - ta)*q*costounitario
  if t==tl:
    dem=demanda()
    if q>=dem:
      q=q - dem
      #print("Llegada ", "%10.2f"%tl, "Demanda ", dem, q)
      v=v+dem
      ganancia=ganancia + gananciaunitaria*dem
    else:
      nv=nv+dem-q
      ganancia=ganancia + q*gananciaunitaria
      #print("Venta perdida", "%10.2f"%tl, "Demanda ", dem, "Ven....")
      q=0
    tl=tl+random.exponential(l_l)
  else:
    if q<s:
      q=S
    #print("Tiempo de reposicion", "%10.2f"%tr, q)
    tr=tr+rep

print("Ganancia: ", ganancia)
print("Costo: ", costo)
print("Ganancia neta: ", ganancia-costo)
```

```python
from random import gauss, random
from math import log

pe=50000    # pago a personal por quincena
tpe=15      # proximo tiemmpo de pago quincenal
n=10000     # numero de clientes
c=150       # cuota mensual por poliza
lam=1       # tasa de ocurrecia de siniestros
u=35000     # media del pago por siniestro
o=5000      # desviacion estandar de los pagos por siniestro
c0=10000000 # capital inicial
T=365       # tiempo de simulacion en dias
ta=0        # tiempo del proximo siniestro
tp=30       # tiempo del proximo pago de los clientes
Cap=c0      # capital en el tiempo t
t=0         # tiempo actual

while min(ta, tp, tpe)<T:
  t=min(ta, tp, tpe)
  if t==ta:
    Cap=Cap - gauss(u,o)
    print("Siniestro ", "%10.3f"%t, "%10d"%Cap)
    ta=ta - (1/lam)*log(random())
  elif t==tp:   # tiempo del proximo pago de polizas
    Cap=Cap + n*c
    print("Pago de polizas ","    ", "%10.3f"%t, "%10d"%Cap)
    tp=t +30
  else: # tpe es pago de nomina
    Cap=Cap - pe
    print("Pago de nomina ","    ", "%10.3f"%t, "%10d"%Cap)
    tpe=tpe+15
print("Ganancia ", "%15.2f"%(Cap - c0))
print("Ganancia porcentual ", "10.3f"%(100*(Cap - c0)/c0))
# < >
```
