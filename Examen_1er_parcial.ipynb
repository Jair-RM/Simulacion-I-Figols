```python
#  1A) Estime por simulacion raiz(5)
#  1B) Cuantas iteraciones debe realizar para asegurar que la
#     probabilidad de que el error sea mayor que 0.01, sea
#     menor que 01?
print("El valor real es de :", sqrt(5)) #Valor real

N=1000000
n=0
lx=[]
ly=[]
lc=[]
for i in range(N):
  r=3*random()
  if r**2 < 5:
    n=n+1
  if i>10000 and i&1000==0:
    lx.append(i)
    ly.append(3*n/(i+1))
    lc.append(sqrt(5))
x=3*n/N
print("Nuestra estimacion es de:", x)
print("El cuadrado de ese numero es: ", x**2)
plt.plot(lx, ly)
plt.plot(lx, lc)
plt.show()
```

```python
# 2. Sea Z=maximo(U1, U2), U1 Y U2 uniformemente distribuidas en
#    [0,1] e independientes
# a) Estime por simulacion el valor esperado de Z
# b) Estime por simulacion P(Z>0.4)

def Z():
  u1=random()
  u2=random()
  if u1>u2:
    max=u1
  if u2>u1:
    max=u2
  return max

N=100000
l=[]
prob=0
esp=0

for i in range(N):
  m=Z()
  l.append(m)
  if m>0.4:
    prob+=1
  esp+=m

esp=esp/N
prob=prob/N

print("El valor esperdo de Z es: ", esp)
print("P(Z>0.4) = ", prob)

```
