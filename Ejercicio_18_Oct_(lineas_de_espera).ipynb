```python
from random import expovariate, choice
def n_servidores(l_l, l_s, T):
  infi=2*T  ##Basta con que sea mas grande que el tiempo de simulacion
  n=len(l_s)    ##NUmero de servidores
  t=0
  l=0        # numero de personas en el sistema
  tl=0
  s_l=[]   ##servidores desoupados en todo momento
  for i in range(n):
    s_l. append(i)
  ts=[]
  for i in range(n):
    ts.append(infi)
  le=0   # numero de personas esperando
  while min(tl, min(ts)) < T:    ## minino de los proximos tiempos de salida para cada servidor
    t = min(tl, min(ts))
    if t==tl:
      l=l+1
      print("Llegada", "%10.3f"%t, " ", "%4d"%l)
      tl=tl + expovariate(l_l)
      if len(s_l) > 0:                       # s_l servidores vacios
        servidor= choice(s_l)
        print("   Pasa al servidor:   ", servidor)
        ts[servidor]=t + expovariate(l_s[servidor])
        s_l. remove(servidor)
      else:
        print("   En espera")
        le=le + 1
    else:
      l=l+1
      servidor=ts.index(t)
      print("        ", "%10.3f"%t, "sale por el servidor: ", servidor)
      s_l. append(servidor)
      if le > 0 and len(s_l)>0:
        print("   Termina espera pasa al sevidor: ", servidor)
        ts[servidor]=t + expovariate(l_s[servidor])
        le=le - 1
        s_l. remove(servidor)
      else:
        ts[servidor]= infi     ##sale pero no hay gente esperando


    print("*******Servidores vacios ", s_l, le)

n_servidores(0.7, [0.5, 0.4, 0.4], 100)
```
